_______________________________________________________________________________________________________________SQL databases are relational databases. They store information in tables and those tables are linked to each others using keys (foreign/prinary keys). They are really useful when dealing with related/structured data. On the other hand, NoSQL databases are a more flexible way to store data with no specific format/structure. They are faster at retrieving data in an chronological order. A good use example is the storing of chat messages where you never now how much message you gonna store and the message format. Data can be structured by key-value pair (MongoDB) or key multi-value pair (Amazon RDS). Also they are easier to split in distinct set allowing you to use the power of distributed computation to execute requests on Resilient Distributed Datasets(Big Data like Hadoop/Spark)
